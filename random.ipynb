```python
import requests
import json

# Get session results for top 10 positions
session_key = 9904  # You can change this to any session key
url = f"https://api.openf1.org/v1/session_result?session_key={session_key}&position<=10"
response = requests.get(url)
session_results = response.json()

# Sort by position to ensure proper order
session_results.sort(key=lambda x: int(x['position']))

print(f"Top 10 drivers for session {session_key}:")
print("-" * 40)
for result in session_results:
    position = result['position']
    driver_number = result['driver_number']
    print(f"Position {position}: Driver #{driver_number}")
```
